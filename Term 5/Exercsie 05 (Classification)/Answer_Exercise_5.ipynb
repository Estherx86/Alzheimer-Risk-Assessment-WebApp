        <li>
                داده های پرت (Outlier) هارا به روش IQR حذف کنید
        </li>
        <li>
            <font color="red">نکته مهم:</font> استفاده از مقادیر پیشفرض Q1 و Q3 داده پیشنهاد نمی شود. این دیتاست داده های کمی دارد و حذف تعداد زیادی از داده ها توصیه نمی شود. شما می توانید از مقادیر کمتری/بیشتری برای Q1 و Q3 استفاده کنید. مثلا مقدار 0.20 به جای 0.25 برای Q1 و مقدار 0.80 برای Q3.(مقادیر را خودتان انتخاب کنید)
        </li>
    </font>
</div>
<br>

```python
def remove_outliers_iqr(df, q1=0.05, q3=0.95):
    # انتخاب ستون‌های عددی
    numeric_cols = df.select_dtypes(include=['int64', 'float64']).columns
    df_cleaned = df.copy()      # کپی از دیتافریم اصلی
    original_count = len(df)    # تعداد داده‌های اولیه

    for column in numeric_cols:
        Q1 = df[column].quantile(q1)
        Q3 = df[column].quantile(q3)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # فیلتر کردن داده‌ها برای این ستون
        df_cleaned = df_cleaned[(df_cleaned[column] >= lower_bound) &
                                (df_cleaned[column] <= upper_bound)]

    # تعداد داده‌ها بعد از حذف
    cleaned_count = len(df_cleaned)
    outliers_removed = original_count - cleaned_count

    print(f"Original Size: {original_count}")
    print(f"Removed Outliers: {outliers_removed}")
    print(f"Remained Data: {cleaned_count}")
    print(f"Evaluated Columns: {list(numeric_cols)}")

    return df_cleaned


df_cleaned = remove_outliers_iqr(df)
```

```python
# TODO: KMeans  پیدا کردن تعداد کلاستر های مناسب

def elbow_method(X, max_k=10):
    """
    پیاده سازی Elbow Method برای یافتن تعداد بهینه خوشه‌ها در K-Means

    پارامترها:
    - X: دیتاست عددی (numpy array یا pandas DataFrame)
    - max_k: حداکثر تعداد خوشه‌هایی که می‌خواهیم تست کنیم (پیش‌فرض: 10)

    خروجی:
    - نمودار Elbow برای انتخاب بهترین مقدار K
    """
    # لیست برای ذخیره WCSS
    wcss = []

    # محاسبه WCSS برای k های مختلف
    for k in range(1, max_k + 1):
        kmeans = KMeans(n_clusters=k,
                        random_state=42,
                        init="k-means++",
                        n_init=20)
        kmeans.fit(X)
        wcss.append(kmeans.inertia_)  # WCSS ذخیره

    # رسم نمودار Elbow
    plt.figure(figsize=(8, 5))
    plt.plot(range(1, max_k + 1), wcss, marker='o', linestyle='--')
    plt.xlabel("Number of Clusters (K)")
    plt.ylabel("WCSS")
    plt.title("Elbow Method for Optimal K", fontsize=14, color='darkblue')
    plt.xticks(range(1, max_k + 1))
    plt.grid(True)
    plt.show()

elbow_method(df_scaled)
```

```python
# TODO :  KMeans پیاده سازی
def perform_kmeans_clustering(df_pca, k=4, df_cleaned=None):
    """
    اجرای K-Means بر روی دیتاست و افزودن لیبل خوشه‌ها به DataFrame اصلی (df_cleaned)

    پارامترها:
    - df_pca: دیتافریم شامل داده‌های کاهش یافته (مثلاً PCA)
    - k: تعداد خوشه‌ها (پیش‌فرض: 4)
    - df_cleaned: دیتافریم اصلی که به آن خوشه‌ها اضافه می‌شود

    خروجی:
    - df_cleaned: دیتافریم با یک ستون جدید به نام 'Cluster_KMeans' که خوشه هر نمونه را مشخص می‌کند.
    """
    # اگر df_cleaned ورودی نباشد، یک کپی از df_pca ایجاد کن
    if df_cleaned is None:
        df_cleaned = df_pca.copy()

    # اجرای K-Means
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=50)
    clusters = kmeans.fit_predict(df_pca)

    # افزودن برچسب خوشه‌ها به DataFrame اصلی
    df_cleaned['Cluster_KMeans'] = clusters

    return df_cleaned
```

```python
# TODO: GMM پیدا کردن تعداد کلاستر های مناسب
def gmm_bic_method(X, max_k=10):
    """
    پیاده‌سازی BIC برای یافتن تعداد بهینه خوشه‌ها در GMM

    پارامترها:
    - X: دیتاست عددی
    - max_k: حداکثر تعداد خوشه‌هایی که می‌خواهیم تست کنیم

    خروجی:
    - نمودار BIC برای انتخاب بهترین مقدار K
    """
    bic_scores = []

    # محاسبه BIC برای k های مختلف
    for k in range(1, max_k + 1):
        gmm = GaussianMixture(n_components=k, random_state=42)
        gmm.fit(X)
        bic_scores.append(gmm.bic(X))

    # رسم نمودار BIC
    plt.figure(figsize=(8, 5))
    plt.plot(range(1, max_k + 1), bic_scores, marker='o', linestyle='--')
    plt.xlabel("Number of Clusters (K)")
    plt.ylabel("BIC")
    plt.title("BIC Method for Optimal K in GMM", fontsize=14, color='darkblue')
    plt.xticks(range(1, max_k + 1))
    plt.grid(True)
    plt.show()

# رسم نمودار BIC
gmm_bic_method(df_pca)
```

```python
#TODO: GMM پیاده سازی
def perform_gmm_clustering(df_pca, k=4, df_cleaned=None):
    """
    اجرای GMM بر روی دیتاست و افزودن لیبل خوشه‌ها به DataFrame اصلی (df_cleaned)

    پارامترها:
    - df_pca: دیتافریم شامل داده‌های کاهش یافته (مثلاً PCA)
    - k: تعداد خوشه‌ها (پیش‌فرض: 4)
    - df_cleaned: دیتافریم اصلی که به آن خوشه‌ها اضافه می‌شود

    خروجی:
    - df_cleaned: دیتافریم با یک ستون جدید به نام 'Cluster_GMM' که خوشه هر نمونه را مشخص می‌کند.
    """
    # اگر df_cleaned ورودی نباشد، یک کپی از df_pca ایجاد کن
    if df_cleaned is None:
        df_cleaned = df_pca.copy()

    # اجرای GMM
    gmm = GaussianMixture(n_components=k, random_state=42)
    clusters = gmm.fit_predict(df_pca)

    # افزودن برچسب خوشه‌ها به DataFrame اصلی
    df_cleaned['Cluster_GMM'] = clusters

    return df_cleaned
```
